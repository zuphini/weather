```python
import requests
import pip
from requests import get

if int(pip.__version__.split('.')[0])>9:
    from pip._internal import main
else:
    from pip import main
def install(package):
    main(['install', package])

install('BeautifulSoup4')
install('pandas')
install('numpy')

from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
## Grabbing the webpage and parsing the html with bs4

page = requests.get("https://weather.com/weather/tenday/l/19ceedf6cc8356209c7991c622ca03efe210636c0516d4b2ec9a2d6f1996dabf")
soup = BeautifulSoup(page.content, "html.parser")
```

```python
## Creating lists from the different data sources using list comprehensions

days = [x.text for x in soup.find_all(attrs={"class": "DailyContent--daypartName--3G5Y8"})] 

temps = [x.text for x in soup.find_all(attrs={"class": "DailyContent--temp--_8DL5"})] 

infos = [x.text for x in soup.find_all(attrs={"class": "DailyContent--narrative--3AcXd"})]

humidsnwinds = [x.text for x in soup.find_all(attrs={"class": "DailyContent--value--3Xvjn"})]

humids = humidsnwinds[0::2]

winds = humidsnwinds[1::2]

## The direct class for the wind data
##winds = [x.text for x in soup.find_all(attrs={"class": "Wind--windWrapper--1Va1P"})]
```

```python
## Creating a dictionary of the dataset

data = {'Temp': temps,
        'Humidity': humids,
        'Wind': winds,
        'Narrative': infos}    
```

```python
## Creating the Dataframe with pandas

pd.DataFrame(data, index=days)
```
